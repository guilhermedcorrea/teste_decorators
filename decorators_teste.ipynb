```python


def add_binomial_probability(n, p):
    def decorator(func):
        def wrapper(*args, **kwargs):
            result = func(*args, **kwargs)
            
            if isinstance(result, pd.Series):
                values = result
                probabilities = binom.pmf(values, n, p)
                result_prob = pd.Series(probabilities, index=result.index, name=f'{result.name}_prob')
            
                return result_prob
            else:
                return result
        
        return wrapper
    return decorator

@add_binomial_probability(n=10, p=0.5)
def calculate_binomial_function(series):
    return series


processed_series = calculate_binomial_function(data['statusv'])
print(processed_series)
```

```python


def preprocess_pca_train_test(data, target_col, test_size=0.2, random_state=None):
    X = data.drop(columns=[target_col])
    y = data[target_col]
    
    pca = PCA(n_components=0.95)  # Preserve 95% of variance
    X_pca = pca.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=test_size, random_state=random_state)
    
    return X_train, X_test, y_train, y_test

def one_hot_encode_categoricals(func):
    """Converte Categoricas"""
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        
        if isinstance(result, pd.DataFrame):
            categorical_columns = result.select_dtypes(include=['object']).columns
            
            for column in categorical_columns:
                if result[column].isnull().any():
                    result[column].fillna('Naotem', inplace=True)
                result[column] = result[column].astype(str) 
            
            encoder = OneHotEncoder(drop='first')
            encoded_data = encoder.fit_transform(result[categorical_columns])
            encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(categorical_columns))
            
            result = pd.concat([result, encoded_df], axis=1)
            result.drop(columns=categorical_columns, inplace=True)
        
        return result
    return wrapper

@one_hot_encode_categoricals
def preprocess_data(data):

    return data

preprocessed_data = preprocess_data(data)
print("Preprocessed Data:")
print(preprocessed_data)

X_train, X_test, y_train, y_test = preprocess_pca_train_test(preprocessed_data, target_col='valores')
print("\nPCA Processed Data:")
print("X_train:", X_train)
print("X_test:", X_test)
print("y_train:", y_train)
print("y_test:", y_test)


X_train



```
